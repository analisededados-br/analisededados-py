# Filtro

```python
tseFiltrado = tse[['NM_CANDIDATO','DS_CARGO','DS_SIT_TOT_TURNO', 'SG_PARTIDO', 'DS_GENERO', 'DS_COR_RACA', 'DS_GRAU_INSTRUCAO', 'DS_OCUPACAO', 'DS_ESTADO_CIVIL', 'SG_UF']]

tseFiltrado = tseFiltrado[(tseFiltrado['DS_CARGO'] == 'DEPUTADO FEDERAL') & (tseFiltrado['DS_SIT_TOT_TURNO'].str.startswith('ELEITO'))]


tseFiltrado
```

```python
deputadosMG = tseFiltrado[(tseFiltrado['SG_UF'] == 'MG')]
deputadosMG = deputadosMG[['SG_PARTIDO', 'DS_GENERO', 'DS_COR_RACA', 'DS_GRAU_INSTRUCAO', 'DS_OCUPACAO', 'DS_ESTADO_CIVIL']]
```

# Moda

```python

deputadosMG['new'] = 'a'
info_geral = deputadosMG.groupby('new').agg(lambda x: x.value_counts().index[0] + f"\n({x.value_counts().iloc[0]} || {round(x.value_counts().iloc[0]/len(x)*100,2)}%)")
info_geral.transpose()
```
